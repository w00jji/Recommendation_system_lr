```python
df_filled = df.copy()

for i in range(len(df_filled)):
    mean = np.nanmean(df_filled[i])
    # print(mean)
    df_filled[i] = np.nan_to_num(df_filled[i], nan=mean)
    # print(df_filled[i])
    df_filled[i] -= mean

df_filled
```

```python
from sklearn.metrics.pairwise import euclidean_distances # 유클리드 거리 계산

distance_matrix=euclidean_distances(df_filled) # 유클리드 거리계산
distance_matrix
```

#### 코사인 유사도, 유클리드 거리 차이

- 코사인 유사도 : 두 벡터 간의 각도를 계산하여 유사성을 측정 , 두 유저가 영화를 동일한 비율로 좋아하더라도, 그 평점의 절대적 크기는 무시 -> 평점 패턴(방향성)에만 초점이 맞춰짐

    - 예시 : 한 유저가 5점, 다른 유저가 4점을 주었더라도 둘의 평점 패턴이 유사하면 코사인 유사도는 높게 나옴

    - 장점 : 다른 유저들과의 영화 평점의 유사한 패턴을 찾기 좋다 , 절대적인 평점이 다르더라도 영화 취향이 비슷하다면 추천에 유리


- 유클리드 거리 : 두 유저 간의 평점 차이를 기반으로 유사성을 측정. 평점간 차이를 직접적으로 계산하기에 절대적 평점 크기가 중요  -> 유사한 점수를 줬다면 유사도가 높고, 큰차이가 나면 유사도가 낮음

    - 장점 : 평점의 절대적인 차이를 고려하기에 , 유저 간에 실제로 얼마나 다른 점수를 줬는지 정확하게 반영함

#### 결과가 다른 이유

- 코사인 유사도 : 평점 패턴이 비슷한 유저에게 높은 유사도를 줌 -> 절대적 평점 차이 무시 , 즉, 평점 크기가 달라도 유사한 영화 선호 경향이 있으면 추천할 가능성 높음

- 유클리드 거리 : 평점 차이를 직접적으로 고려하기에 , 비슷한 점수를 준 유저들을 기반으로 추천 , 즉, 절대적인 평점 차이가 큰 경우에 유사도가 낮게 평가됨

#### 두 방식 차이 분석

- 코사인 유사도 방식의 추천: 다양한 평점 패턴을 반영하여 사용자가 보지 않은 영화들에 대해 유사한 취향을 가진 사람들이 추천한 영화를 제공.

- 유클리드 거리 방식의 추천: 절대적인 점수 차이를 반영하여, 실제로 사용자가 비슷한 점수를 줄 만한 영화를 추천.

#### 개선사항

-코사인 유사도, 유클리드 거리의 각각의 장점을 활용하여 두 결과를 결합한 모델


- 코사인 유사도 : 평점 패턴 잘 반영 but 절대적 차이 무시
- 유클리드 거리 : 절대적 평점 차이 잘 고려

결론


코사인 유사도는 취향의 패턴을 반영하므로, 사용자의 평점 성향을 반영한 영화를 더 잘 추천


유클리드 거리는 절대적인 평점을 더 많이 고려하므로, 사용자가 실제로 높은 점수를 줄 가능성이 높은 영화를 추천



#### 개선 코드

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity

# 1. 유클리드 거리 계산
def calc_euclidean_distances(ratings):
    filled_ratings = ratings.copy()  # 원본 데이터 복사
    for i in range(len(filled_ratings)):
        mean = np.nanmean(filled_ratings[i])  # 각 행의 평균 계산
        filled_ratings[i] = np.nan_to_num(filled_ratings[i], nan=mean)  # NaN을 평균으로 대체
        filled_ratings[i] -= mean  # 평균 빼기
    return euclidean_distances(filled_ratings)

# 2. 코사인 유사도 계산
def calc_cosine_similarity(ratings):
    filled_ratings = ratings.copy()  # 원본 데이터 복사
    for i in range(len(filled_ratings)):
        mean = np.nanmean(filled_ratings[i])  # 각 행의 평균 계산
        filled_ratings[i] = np.nan_to_num(filled_ratings[i], nan=mean)  # NaN을 평균으로 대체
        filled_ratings[i] -= mean  # 평균 빼기
    return cosine_similarity(filled_ratings)
```

```python
# 3. 유사도 결합
def combine_similarity(euclidean_distances, cosine_similarity):
    euclidean_similarity = 1 / (1 + euclidean_distances)
    combined_similarity = (euclidean_similarity + cosine_similarity) / 2
    return combined_similarity
```

```python
# 4. 사용자별 추천
def recommend_movies(ratings, user, k=2, top_n=5):
    euclidean_dists = calc_euclidean_distances(ratings)
    cosine_sims = calc_cosine_similarity(ratings)
    combined_sims = combine_similarity(euclidean_dists, cosine_sims)

    # 유사한 이웃 찾기
    neighbors = np.argsort(-combined_sims[user])[1:k+1]
    user_ratings = ratings[user]
    unrated_movies = np.where(np.isnan(user_ratings))[0]

    predictions = {}

    for movie in unrated_movies:
        neighbor_ratings = ratings[neighbors, movie]
        valid_ratings = neighbor_ratings[~np.isnan(neighbor_ratings)]
        if len(valid_ratings) > 0:
            predictions[movie] = np.mean(valid_ratings)
        else:
            predictions[movie] = np.nan

    sorted_predictions = sorted(predictions.items(), key=lambda x: x[1], reverse=True)
    return sorted_predictions[:top_n]
```

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity
import pandas as pd

# 평점 데이터 로드
ratings = pd.read_csv('/content/drive/MyDrive/IMbank_추천시스템/추천시스템 실습 데이터/data/ratings.csv')

# 영화 데이터 로드
movies = pd.read_csv('/content/drive/MyDrive/IMbank_추천시스템/추천시스템 실습 데이터/data/movies.csv')

# 영화 ID와 제목의 매핑 생성
movie_names = {}
for i in range(len(movies)):
    a = movies.iloc[i]
    movie_names[a['movieId']] = a['title']






```

```python

# 추천 실행 (예시로 0번 사용자를 위한 추천)
user = 0
top_n = 5  # 상위 5개 추천
recommendations = recommend_movies(ratings.values, user, top_n=top_n)
```
