## ratings 데이터로 행렬 인수분해를 통해서 누락값들을 예측하고 영화 추천!

```python
import pandas as pd

ratings = pd.read_csv('/content/drive/MyDrive/IMbank_추천시스템/추천시스템 실습 데이터/data/ratings.csv')
ratings.head()
```

```python
import random

# 행렬 인수분해를 위한 초기화
k = 5  # 잠재 요인의 수
m, n = ratings_pivot.shape
U = np.random.rand(m, k)  # 사용자 잠재 행렬
V = np.random.rand(n, k)  # 영화 잠재 행렬
```

```python
import numpy as np

# 하이퍼파라미터
m, n, k = 1000, 500, 10  # 예시로 사용자 수, 항목 수, 잠재 요인 수 설정
lr = 0.0001  # 학습률 조정
lambda_reg = 0.1  # 정규화 하이퍼파라미터

# 초기 행렬 생성 (더 작은 값으로 초기화)
U = np.random.rand(m, k) * 0.001  # 사용자 잠재 요인 행렬
V = np.random.rand(n, k) * 0.001  # 항목 잠재 요인 행렬

# 가상의 평점 행렬 생성 (예시로 0~5 사이의 값)
ratings_pivot = np.random.randint(0, 6, size=(m, n)).astype(float)

# 마스크 생성 (결측값을 처리하기 위한 마스크)
mask = np.random.rand(m, n) > 0.5  # 50%의 결측값 생성
ratings_pivot[~mask] = np.nan  # 마스크에 따라 NaN 값 할당

# 행렬 인수분해
for i in range(5000):
    # 예측 평점 계산
    ratings_pred = U @ V.T

    # 오차 마스크 설정
    E = np.zeros((m, n))
    E[mask] = ratings_pred[mask] - ratings_pivot[mask]

    # 경량 조정
    U_grad = E @ V
    V_grad = E.T @ U

    U -= U_grad * lr
    V -= V_grad * lr

    # 값 제한
    U = np.clip(U, -1, 1)  # U 값 제한
    V = np.clip(V, -1, 1)  # V 값 제한

    # 손실 계산 (정규화 추가)
    loss = np.sum(E[mask] ** 2) + lambda_reg * (np.sum(U**2) + np.sum(V**2))

    # 손실 출력 (필요 시)
    if i % 500 == 0:  # 500 에폭마다 손실 출력
        print(f"Epoch {i}, Loss: {loss:.4f}")
```

```python
# 예측된 평점
ratings_pred = U @ V.T

# 결측값을 예측한 평점으로 채우기
ratings_filled = ratings_pivot.copy()  # ratings_pivot이 NumPy 배열이라면
ratings_filled[~mask] = ratings_pred[~mask]
```

```python
# 추천 시스템
for user in range(ratings_filled.shape[0]):  # 사용자 수 만큼 반복
    user_ratings = ratings_filled[user]  # 특정 사용자의 평점 배열
    recommended_movies = np.where(user_ratings >= 4)[0].tolist()  # 평점이 4 이상인 영화의 인덱스

    if recommended_movies:
        print(f'{user + 1}번째 유저, 추천할 영화 IDs: {recommended_movies}')
    else:
        print(f'{user + 1}번째 유저에게 추천할 영화가 없습니다.')
```
